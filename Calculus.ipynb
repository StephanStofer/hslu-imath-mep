# Multivariable Calculus

## Graphing
To draw the Graph of a real valued function of two variables $z=f(x,y)$ use `graph(f)`

```octave
function graph (f)
    ticks = [-3.0:0.1:3.0];
    [x, y] = meshgrid(ticks, ticks);
    z = f;
    surf(x,y,z); 
    title("Graph of function f"); 
endfunction
```

```octave
f = x.^2/2 + y.^2;
graph(f);
```

## Contour Lines
Draw a Contour Line

```octave
x = linspace(-2 * pi,2 * pi);
y = linspace(0,4 * pi);
[X,Y] = meshgrid(x,y);
Z = X.^2-Y.^2;
contour(X,Y,Z)
```

```octave
colormap(jet)
ticks = [-3.0:0.1:3.0];
[x, y] = meshgrid(ticks, ticks);
z = x.^2/2 + y.^2;
contour(x,y,z);
title('Elliptic Bowle using meshgrid')
```
