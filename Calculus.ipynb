# Multivariable Calculus

## Graphing
To draw the Graph of a real valued function of two variables $z=f(x,y)$ use `graph(f)`

```octave
function graph(f)
    ticks = [-3.0:0.1:3.0];
    [x, y] = meshgrid(ticks, ticks);
    z = f;
    surf(x,y,z); 
    title("Graph of function f"); 
endfunction

f = @(x.^(2/2) + (y.^2));
graph(f)
```

```octave
ticks = [-3.0:0.1:3.0];
[x, y] = meshgrid(ticks, ticks);
z = f;
surf(x,y,z); 
title("Graph of function f"); 

```

## Contour Lines
Draw a Contour Line

```octave
x = linspace(-6, 6);
y = linspace(-6, 6);
[X,Y] = meshgrid(x,y);
Z = X.^2-Y.^2;
contour(X,Y,Z)
```

```octave
x = linspace(0, 6);
y = linspace(-3, 3);
[X,Y] = meshgrid(x,y);
Z = 2.*X+3.*Y+1;
contour(X,Y,Z)
```

```octave
colormap(jet)
ticks = [-3.0:0.1:3.0];
[x, y] = meshgrid(ticks, ticks);
z = x.^2/2 + y.^2;
contour(x,y,z);
title('Elliptic Bowle using meshgrid')
```

```octave
y = linspace(0,4 * pi);
[X,Y] = meshgrid(x,y);
Z = X.^2-Y.^2;

figure
contour(X,Y,Z)
```
